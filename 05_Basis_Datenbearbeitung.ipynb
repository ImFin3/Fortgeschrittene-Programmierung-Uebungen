## Aufgabe 1: Slicing
### a) auf Listen
- erstelle eine Liste `slice_list` 
- mit den Werten von 1 bis 30, erstelle die Liste `nicht` manuell

```python
slice_list = []
for i in range(1,31):
    slice_list.append(i)
print(slice_list)
```

- Verwenden die Slicing-Notation, um jedes `zweite` Element aus dieser Liste auszuwählen und in einer neuen Liste zu speichern. 
- Gebe sowohl die ursprüngliche Liste als auch die neue Liste aus.

```python
new_list = slice_list[::2]
print(new_list)
```

- Verwenden die Slicing-Notation mit dem `Startindex 1` und der `Schrittweite 2`, um eine neue Liste namens jedes_zweite_element_ab_index_eins zu erstellen
- (gebe edes `zweite` Element startend mit dem `zweiten` Element)

```python
every_second_element = slice_list[1::2]
print(every_second_element)
```

- Verwenden die Slicing-Notation und selektiere auf den Bereich von den Elementen `5-13` 
- und gebe nur jedes `dritte` Element in der neuen Liste aus

```python
print(slice_list[4:13:3])
```

- gebe mittels Slicing-Notation nur die letzten `vier` Elemente aus

```python
print(slice_list[-4::])
```

- Verwende die Slicing-Notation. Das Slicing soll beim `fünften` Element starten und beim `vorletzen` Element der  Liste enden
- und zeige nur jedes `vierte` Element an in der neuen Liste

```python
print(slice_list[4:-1:4])
```

### c) Slice auf String
- erstelle den String Satz: `Das ist ein langer Satz zum Testen für das slicen in Strings.`

```python
sentence = "Das ist ein langer Satz zum Testen für das slicen in Strings."
```

gebe aus dem String Satz folgendes aus
- Das erste Ausgegebene Element ist das `8` Element 
- Es soll jedes `dritte` Element ausgegeben werden
- Die Liste soll bis auf das `vorletzte` Element geslicet werden
<br> <br> **Lösung** 'e nra meef sleiSis'

```python
print(sentence[8:-1:3])
```

- gebe aus dem String Satz das Wort `Satz` aus
- um das Wort zu bestimmen, sollst du nicht manuell Zählen. Eine Idee ist das verwenden eines for loops
- wenn du jedem Stringelement einen Index zu geordnet hast, kannst du das Wort `Satz` mittels Slicen ausgeben.

```python
start_index = 0
for i in range(len(sentence)):
    if sentence[i:i+4] == "Satz":
        start_index = i

print(sentence[start_index:start_index+4])
```

## Aufgabe 2: allgemeine Datenbearbeitung
### a) erweiteren und bearbeiten des DataFrames aus der Aufgabe `1d)`
- reduziere den DataFrame `df_slice` auf die Columns `A` und `B`, 
- sowie die ersten `10` Rows und nenne den neuen DataFrame `work_df`

```python
work_df = df_slice.loc[0:9, "A":"B"]
work_df
```

- nenne die Columns im Datframes `work_df` um:
    - `A` in `VALUE_1` und 
    - `B` in `VALUE_2`

```python
work_df = work_df.rename(columns={"A":"VALUE_1", "B":"VALUE_2"})
work_df
```

- füge dem Dataframe `work_df` eine neue Column hinzu
- die Column soll den Namen `GERADE_DIV` haben
- die Column `GERADE_DIV` besteht nur aus dem Wert `2`

```python
work_df["GERADE_DIV"] = 2
work_df
```

- erweitere den DataFrame `work_df` um eine neue Column
- kopiere die Column `VALUE_1` mit allen seinen Werten in eine neue Column mit dem Namen `COPY_VALUE_1` 

```python
work_df["COPY_VALUE_1"] = work_df["VALUE_1"]
work_df
```

löschen vom Columns 
- lösche die Column `VALUE_1` aus dem Dataframe `work_df`

```python
work_df = work_df.drop("VALUE_1", axis=1)
work_df
```

erstellen einer neuen berechneten Column in den DataFrame
- die neue Column soll den Wert von `COPY_VALUE_1` mit dem Wert von `VALUE_2` subtrahieren
- schreibe das Ergebnis in eine neue Column mit den Namen `SUB_ERGEBNIS` in den Dataframe `work_df`

```python
work_df["SUB_ERGEBNIS"] = work_df["COPY_VALUE_1"] - work_df["VALUE_2"]
work_df
```

- entferne nun alle duplicate in dem ganzen Dataframe `work_df` 
<br><br> **ACHTUNG** `auch wenn dein work_df keine Doubletten hat`

```python
work_df = work_df.drop_duplicates()
work_df
```

- entferne nun alle duplicate, auf den Columns `VALUE_2` und `GERADE_DIV` aus dem ganzen Dataframe `work_df` 
- behalte beim löschen den `letzten` Eintrag, von allen gleichen duplizierenden Einträgen

```python
work_df = work_df.drop_duplicates(
    subset=["VALUE_2", "GERADE_DIV"],
    keep="last"
)
work_df
```

- erstelle für den Dataframe `work_df` einen neuen `Index` beginnend bei 0

```python
work_df = work_df.reset_index(drop=True)
work_df
```

- erstelle die Summe (über alle Columns) in einer neuen Column mit dem Namen `SUM` in Dataftrane `work_df`

```python
work_df["SUM"] = sum([work_df["VALUE_2"], work_df["GERADE_DIV"], work_df["COPY_VALUE_1"], work_df["SUB_ERGEBNIS"]])
work_df
```

- gebe für die  Column `SUM` im Dataftrane `work_df` jeweils den `minmalen` Wert und den `maximalen` Wert aus

```python
print(work_df["SUM"].min())
print(work_df["SUM"].max())
```

- gebe die Mittelwerte für `alle` Columns aus dem Dataframe `work_df`

```python
print(work_df.median())
```

veränder den Index erneut
- Die Column `SUM` soll der neue Index im Dataftrane `work_df` sein
- Des Weiteren soll die Column `index` gelöscht werden

```python
work_df = work_df.set_index("SUM")
work_df
```

- erstelle eine neue Column namens `ROUND` im Dataftrane `work_df`
- in der Columns `ROUND` sollen die Column `VALUE_2` durch `COPY_VALUE_1` dividiert werden und 
- das Ergebnis soll auf 3 `Nachkommastellen` gerundet werden

```python
work_df["ROUND"] = round(work_df["VALUE_2"] / work_df["COPY_VALUE_1"], 3)
work_df
```

- erstelle eine neue Column namens `POTENZ` im Dataftrane `work_df`
- in der Columns `POTENZ` sollen die jeweiligen Potenzen von `VALUE_2` auf `COPY_VALUE_1` berechnet werden
- das Ergebnis soll hierbei auf 1 Stelle `vor` dem Komma gerundet werden.

```python
work_df["POTENZ"] = round(work_df["VALUE_2"] ** work_df["COPY_VALUE_1"], 1)
work_df
```

- zähle und gebe alle verschiedenen Ausprägungen von `COPY_VALUE_1` im Dataftrane `work_df` aus
<br><br> **value_counts**

```python
work_df["COPY_VALUE_1"].value_counts()
```

- gebe den Dataftrane `work_df` gedreht aus

```python
work_df.transpose()
```
