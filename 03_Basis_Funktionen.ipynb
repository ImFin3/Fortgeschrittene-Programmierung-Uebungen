## Aufgabe 1: Schreibe eine eigene Funktionen
### a) Produkt-Funktion erstellen
- Erstelle eine Funktion, welche das Produkt (also alle Werte miteinander multipliziert) die in einer Liste mitgegeben werden. In der Liste müssen mindestens `3` Werte/Zahlen enthalten sein 
- Aus der Funktion soll zum einen das `Ergebnis`, als auch den Input mit print ausgegeben werden
- Beispiel: Inputliste: 1, 2, 3

```python
num_list = [1,7,3]

def multiply_list(numbers: list):
    result = 1
    for i in numbers:
        result *= i
    print(f"Das Produkt aus {numbers} = {result}")

multiply_list(num_list)
```

### b) Summe-Funktion
- Erstelle eine Funktion, welche die Summe aus `4` "separaten" *Input-Variablen* von den Werten berechnet
- Wenn in der Funktion ein Wert z.B. nicht mit angegeben wird, wird dieser immer mit 0 in die Funktion berechnet 
    - **Tipp** gebe dies im Funktionskopf mit an

### c) Hinzufügen von Werten in die Liste
Daten aus verschiedenen Quellen in einer einzigen Liste zusammenführen müssen. Manchmal erhalten Sie einzelne Datenpunkte, und manchmal erhalten Sie Daten in Form von Listen oder Tupeln. Die Aufgabe ist es, eine Python-Funktion zu schreiben, welche diese Aufgabe effizient erledigt. Die Funktion soll eine vorhandene Liste nehmen und neue Daten hinzufügen, unabhängig davon, ob es sich um einen einzelnen Wert oder eine Gruppe von Werten handelt.
- füge in einer Funktion eine `beliebige Anzahl an Elementen` hinzu:
    - diese können z.B. sowohl in einer `Liste, Tuple als auch in einzelnen Werten` übergeben werden
    - Parameter im Funktionskopf sollten `appendable_list` und `append_values` sein 
- Step 1:
    - Schriebe eine Funktion, welche prüft ob es sich um ein Integer oder einen String handelt
    - Prüfe weiter, ob es sich um eine verkettete Liste oder ein Tupel handelt
- Step 2:
    - Füge die Werte einer Liste hinzu
    - Gebe die neue Liste aus
<br> 
Tipp: 
- nutze `append`
- exemplarhafte Eingabe für die untengezeigte Lösung:
    - []
    - [1, 2]
    - (3, 4)
    - 5
    - 'str'

 



```python
my_list = []

def append_to_list(numbers: list, value):
    if type(value) == str:
        print(f"Beim Wert {value} handelt es sich um einen String")
        numbers.append(value)
    elif type(value) == int:
        print(f"Beim Wert {value} handelt es sich um einen Integer")
        numbers.append(value)
    elif type(value) == list:
        print(f"Beim Wert {value} handelt es sich um eine Liste")
        for element in value:
            numbers.append(element)

print(my_list)
append_to_list(my_list, [])
print(my_list)
append_to_list(my_list, [1,2])
print(my_list)
append_to_list(my_list, [3,4])
print(my_list)
append_to_list(my_list, 5)
print(my_list)
append_to_list(my_list, "str")
print(my_list)

```

### d) Erstelle einen Funktion, welche sich mit sich selbst multipliziert
- Die Funktion soll zum einen die Bedingung prüfen ob der `Eingabewert` ein `int oder float` ist
- Wenn der Eingabewert der Bedingung nicht erfüllt ist (z.B. ein String geliefert wird) soll ein Fahler ausgegeben werden. 
<br><br>**Tipp:** 
<br>raise ValueError ('Falscher Datentyp der Inputvariable')

```python
def mulitply_itself(value) -> int | float:
    if type(value) == int:
        print(f"{value} ist ein Integer!")
        return value*value
    elif type(value) == float:
        print(f"{value} ist ein Float!")
        return value*value
    else:
        raise ValueError(f"\"{value}\" ist weder Integer noch Float!")

print(mulitply_itself(10))
#print(mulitply_itself("asd"))

```

### b) Erstelle einen Code, welcher eine Liste von Zahlen verarbeiten kann. Der Code erstellt wiederrum zwei neue Listen: 
- Eine enthält die `Absolutwerte` der ursprünglichen Zahlen und
- die andere enthält die `Quadrate` der ursprünglichen Zahlen. 
- Schließlich berechnet und gibt der Code die Summe der Werte in jeder dieser neuen Listen aus.

```python
origin_list = [randint(0,10) for _ in range(8)]
square_list = []

for e in origin_list:
    square_list.append(e**2)

print(f"Liste mit Absolutwerten: {origin_list}")
print(f"Liste mit Quadrierten Werten: {square_list}")

print(f"Summe der Absolutwerte: {sum(origin_list)}")
print(f"Summe der Quadrierten Werte: {sum(square_list)}")
```
