# Comprehension


## 1) Comprehension
### a) if-else
- Eine Liste ``number`` mit den Zahlen **1,-2,3,-4,5** erstellen
- Die  Zahlen aus der Liste klassieifzieren, ob diese positive oder negative ist

### b) verschachtelte
- Eine Matrix (Dies entspricht einer Liste von Listen) ``matrix`` mit den Werten **[1, 2], [3, 4], [5, 6]** erstellen
- Die Matrix soll glätted (Flattening) werden. (dies enspricht einer 1-dimensionalen Liste) 

### c) sets
- Eine Liste ``words`` mit den Elementen ***apple, banana, apple, cherry*** soll die länge zählen und die Duplikate entfernen
- **Tipp** benutze die Funktion len

### d) Dictionaries
- Eine Liste ``names`` mit den Namen ***Alice, Bob, Charilie*** sollen die jeweiligen Längen mitgegeben werden.
- Es soll also ein Shlüssel-Wert-Parr erstellt werden

### e) Funktionen mit verwenden
- Eine Liste ``raw_data`` mit den Daten ***["  PIZZA ", "sushi  ", "  TaCoS  "]***
- Entferne aus ``raw_data``  die Leerzeichen (whitespaces) und gebe diese aus

### f) Kombinatorik: Erstellen von Kombinationen (z. B. Kartenspiel: Farbe + Wert)
- In einem Deck gibt es die Farben **Herz, Pik, Kreuz und Caro**, welche in einer Liste  ``farben`` gespeichert sind
- Eine gibt in einem Deck die Farben **Ass und König** welche in einer Liste  ``werte`` gespeichert sind
- gebe nun die möglichen Kombinationen aus
