## Aufgabe 1: schreibe eine Text-datei nach den folgenden Schritten

- importiere die Bibliothek os
- importiere csv
- importiere pandas

```python
import os
import csv
import pandas as pd
```

- hier wird ein neuer `leerer` Ordner erstellt, in dem gleichen Pfad aus dem das Skript läuft
- mittels der os Biliothek,  **mkdir** ist ein Terminalbefehl
- in dem Ordner wird später die erzeugte Datei gespeichert

```python
# create new folder
if not os.path.exists("temp"):
    os.mkdir("temp") 

# create basic text file
text_string = """Moin Moin,
Ostfriesland ist ein schönes Land!
Bis bald"""
```

- erstelle eine `txt` Datei namens `ostfriesland` in den erstellten Ordner `temp`
<br>
<br>**Tipp:** verwende os.path.join 

```python
file_path = os.path.join("temp", "ostfriesland.txt")
```

- schreibe den ganzen String `text_string` in die `ostfriesland` txt. Datei

```python
file = open(file_path, "w")
file.write(text_string)
```

- beende die Schreibfunktion

```python
file.close()
```

- kontrolliere, ob die Datei in deinem Verzeichnis enthalten ist
- kontrolliere, ob die Datei den String `text_string` enthalten ist
- lese `read` die beschriebene Datei `ostfriesland` in diesem Notebook aus

```python
if not os.path.exists(file_path):
    print("File doesnt exist")
else:
    with open(file_path, "r") as file:
        contents = file.read()
        if  contents != text_string:
            print("File doesn't include the text_string")
        else:
            print(contents)
```

## Aufgabe 2: Füge der `ostfriesland` Datei einen zusätzlichen Text hinzu
- lösche **`nicht`** den bestehenden Text
- Erweitere die `ostfriesland` Datei um folgende Einträge, in jeweils einer Zeile
    1. I'm adding more lines
    2. and more...

## Aufgabe 3: Erstelle nun eine csv
- Erstelle eine CSV mit dem Namen `data.csv`
- Die CSV soll aus zwei `Columns` bestehen. Die **Columnnamen** sollen `ID, FARBE` sein
- Anschliessend sollen exemplarisch drei rows zu den Columns hinzugefügt : <br><br>
    - 0. 1, rot
    - 1. 2, blau
    - 2. 3, rot
<br><br> **Tipp**  writer = csv.writer

```python
csv_path = os.path.join("temp", "data.csv")

with open(csv_path, "w", newline="") as file:
    writer = csv.writer(file) # delimiter="\t"

    writer.writerow(["ID", "FARBE"])

    writer.writerow([0, "rot"])
    writer.writerow([1, "blau"])
    writer.writerow([2, "rot"])
```

- lese nun die eingelesenen Daten aus der CSV aus <br><br>
**Tipp:** erstelle als eine leere Liste und füge die einzelnen Elemente der Liste hinzu

```python
daten = []  # leere Liste

with open(csv_path, "r") as file:
    reader = csv.reader(file)
    for row in reader:
        daten.append(row)

# Ausgabe
for zeile in daten:
    print(zeile)
```

- erstelle einen Dataframe aus der vorher erstellten Liste
<br><br> **Tipp** benutze ein for loop zum schreiben in einen Dataframes

```python
header = daten[0]
data = daten[1:]

my_df = pd.DataFrame(data, columns=header)
my_df
```

- Lösche die erstellten Datein 
- lösche den erstellten Ordner
<br><br>
**Tipp** verwende remove und rmdir

```python
os.remove(file_path)
os.remove(csv_path)

os.rmdir("temp")
```
