## Aufgabe 1: 

Durch die zufälligen Werte in dem Dataframes `data_np` werden die angezeigten Lösungen nicht mit euren Lösungen identisch sein. Die angezeigten Lösungen sollen euch nur einen Indikator anzeigen, was bei den Aufgaben gemacht werden soll.

```python
import pandas as pd
import numpy as np
import random
import string

data_np = {'column_int': np.random.randint(1,10, size=100)
        , 'column_int_1': np.random.randint(1,10, size=100)
        , 'column_float': np.random.uniform(low=-100, high=100, size=(100))
       }

data_np = pd.DataFrame(data_np)
data_np.head()
```

- ersetze in dem Dataframes `data_np` in der Column `column_int_1` die Werte `6 und 7`  mit `NaN-Werten`
<br><br>**Tipp** verwende replace und gebe beide Elemente in einer Liste wieder

```python
data_np['column_int_1'] = data_np['column_int_1'].replace([6, 7], np.nan)
data_np
```

- bestimme die Grösse (columns und records) von dem Dataframes `data_np`

```python
data_np.shape
```

- berechne für alle `Column` aus dem Dataframes `data_np` den dazugehörigen mean pro Column

```python
col_list = data_np.columns

for col in col_list:
    print(col + ":",data_np[col].mean())
```

- berechne nun den mean pro `Row` für den Dataframe `data_np`
- Was muss sich ändern in den Parametern von mean?

```python
data_np.mean(axis=1)
```

- füge dem Dataframe `data_np` eine neue `Column` mir dem Namen `mean_row` hinzu 
- Die Column `mean_row` wird der durchschnitts Wert pro row übergeben

```python
data_np["mean_row"] = data_np.mean(axis=1)
data_np
```

- füge dem Dataframe `data_np` eine neue Column mit dem Namen `column_sum` hinzu
- Der Column `column_sum` soll die Summe pro row `mittels apply` übergeben werden

```python
data_np["column_sum"] = data_np.apply(func=lambda row: row.sum(), axis=1)
data_np
```

- füge dem Dataframes `data_np` eine neue row hinzu
- übergebe der neuen row die Summe pro Column mittels loc

```python
sum_row = data_np.sum(axis=0, numeric_only=True)

data_np.loc[len(data_np)] = sum_row
data_np
```
