### g) Strings justieren anhand der Länge
- Gebe die einzelnen Elemente der Liste `string_liste_just` aus?

- Nun sollen die Elemente der `string_liste_just`  **`rechtsbündig`** ausgegeben werden (an dem maximalen Wert)
<br><br>
**Tipps:**
- kreiere eine Variable max_len
- du musst die Ausrichtung mitgeben werden --> Bestimme die maximale Länge der Elemente im Array --> mittels for-loop

### c) Erstelle Spielpaarungen

- Diese Mannschaften sollen jeweils gegen einander spielen.
- erstelle alle verschiedenen möglichen Spielpaarungen (**Tipp: enumerate**) und gebe diese anschliessend aus

```python
teams = ["Norwegen", "Schweden", "Finnland", "Dänemark"]

for i, team1 in enumerate(teams):
    for team2 in teams[i + 1:]:
        print(team1, "vs", team2)
```

### d) Listen anpassen

- Die `liste_a` ist die Basis Liste
- Wenn der Wert von `liste_a` in Liste `liste_b` enthalten ist, soll der Wert aus `liste_b` direkt nach dem gleichen Element in `listen_a` angefügt werden
- Wenn nicht soll der Wert an das Ende der `liste_a` hinzugefügt werden.
<br><br>
**Tipp**
- Verwende ein for-loop und eine if-else-cause
<br><br> Die Lösung sieht wie folgt aus: *['a', 'a', 'a', 'c', 'c', 'e', 'e', 'g', 'g', 'f', 'f', 'd', 'd', 'h']*

```python
liste_a = ['a','c', 'e', 'g']
liste_b = ['c', 'f', 'd', 'h', 'a', 'e', 'd', 'a', 'f', 'g']
```

```python
for e in liste_b:
    if e in liste_a:
        index = liste_a.index(e)
        liste_a.insert(index + 1, e)

print(liste_a)
```
