### a) Wie viele Tage hat ein Monat? -> mit zufälliger Anzahl

Entwerfe ein Code/Programm, welches die Anzahl der Tage für einen zufälligen Monat beschreibt. 
<br> **Die Schaltjahr-Reglung muss ``nicht`` berücksichtigt werden.**
<br> <br> Folgende Schritte für die Umsetzung:

* Step 1: Stelle die Frage: `Welcher Monat ist es?`
* Step 2: Gebe eine `zufällige` Zahl **zwischen 1 und 12** aus
* Step 3: Erstelle die jeweiligen Verzweigungen:
    - die Monate: Januar, März, Mai, Juli, August, Oktober und Dezember haben `31` Tage, 
    - die Monate April, Juni, September und November haben `30` *und*
    - der Februar `28` Tage
* Step 4: Gebe das richtige Ergebnis, inklusiv Monat und Tage aus

```python
def get_month_length(month: int, is_leap_year: bool) -> int:
    months_with_31 = [1, 3, 5, 7, 8, 10, 12]
    months_with_30 = [4, 6, 9, 11]
    moths_with_28 = [2]

    if month in months_with_31:
        return 31
    elif month in months_with_30:
        return 30
    elif month in moths_with_28:
        if is_leap_year:
            return 29
        else:
            return 28

print("Welcher Monat ist es?")
random_number = random.randint(1, 12)
print(random_number)
month_length = get_month_length(random_number, False)
print(f"Der {random_number}. Monat im Jahr hat {month_length} Tage.")
```

### d) Rate die richtige Zahl
- Der Code generiert  eine Zufallszahl zwischen `1 und 30` - diese wird nicht angezeigt
- Schriebe einen Code welche die Zufallszahl eraten kann. Dabei soll es `maximal 5 Versuchen` zum Raten geben
- Folgende Bedingungen / Step muss es geben:
    - die geratene Zahl als manuellen Input eingeben können
    - bei der ersten Eingabe soll `Erster Versuch` stehen
    - wenn die geratene Zahl richtig ist, soll die `Zahl`, die `Anzahl der Versuche` und ein `Glückwusch` ausgegeben werden
    - wenn die Zahl nicht richtig ist:
        - soll für die geratene Zahl ausgeben werden, ob diese zu klein oder zu gross ist im vergleich zur Zufallszahl ist
        - und es soll ein weiter Versuch mit dem Raten gestartet werden. Text `Weitere Versuche`
    - wenn du `mehr als die 5 Versuche` hast soll ein **Fehler** und die Zufallszahl ausgegeben werden

```python
random_number = random.randint(1,30)

print("Errate die Zahl zwischen 1 und 30!")
guessed_number = int(input("Erster Versuch"))
guess_count = 1
if guessed_number != random_number:
    if guessed_number < random_number:
        print(f"Deine geratene Zahl {guessed_number} ist kleiner als die zu suchende Zahl.")
    elif guessed_number > random_number:
        print(f"Deine geratene Zahl {guessed_number} ist größer als die zu suchende Zahl.")
    guess_count += 1

    while True:
        if guess_count > 5:
            print(f"Du hast mehr als 5 versuche gebraucht, die gesuchte Zahl war {random_number}!")
            break
        else:
            guessed_number = int(input(f"{guess_count}. Versuch"))
            if guessed_number == random_number:
                print(f"Glückwunsch! Du hast die Zahl {random_number} in {guess_count} Versuchen herausgefunden!")
                break
            elif guessed_number < random_number:
                print(f"Deine geratene Zahl {guessed_number} ist kleiner als die zu suchende Zahl.")
            elif guessed_number > random_number:
                print(f"Deine geratene Zahl {guessed_number} ist größer als die zu suchende Zahl.")
            guess_count += 1

else:
    print(f"WOW, das ist richtig die zu suchende Zahl war {random_number}!")
```

### e) Wüefelspiel
Erstelle ein Würfelspiel nach folgenden Kriterien:
- `3 Spieler` würfeln eine Zahl zwischen `1 bis 6`
- Der Spieler mit dem `höchsten Wert` in einer Runde bekommt einen Punkt
- Wenn aber `2 Spieler den gleichen Wert` haben, wird in der Runde `kein Punkt` verteilt
- Haben `alle 3 Spieler` die gleiche Zahl, wird jedem Spieler ein Punkt `abgezogen`
- Der Spieler der zuerst `6 Punkte` erreicht hat `gewonnen`
<br>
`Gebe an wie viele Runden der Siegergebraucht hat und die Ergeisse der einzelnen Runden`

```python
class Player:
    def __init__(self, points: int = 0):
        self.points = points
        self.dice_roll = 0

    def roll_dice(self):
        self.dice_roll = random.randint(1, 6)

    def lose_point(self):
        self.points -= 1
        if self.points < 0:
            self.points = 0

    def win_point(self):
        self.points += 1

    def reset(self):
        self.points = 0
        self.dice_roll = 0

class Game:
    def __init__(self):
        self.rounds = 0
        self.player1 = Player()
        self.player2 = Player()
        self.player3 = Player()

    def start_game(self):
        while not self.is_game_over():
            self.rounds += 1
            self.reroll_all_dice()
            self.print_dice_rolls()
            self.distribute_points()
            self.print_points()
            print()
        self.end_game()

    def is_game_over(self) -> bool:
        if self.player1.points < 6 and self.player2.points < 6 and self.player3.points < 6:
            return False
        else:
            return True

    def reroll_all_dice(self):
        self.player1.roll_dice()
        self.player2.roll_dice()
        self.player3.roll_dice()

    def print_dice_rolls(self):
        print(f"Runde {self.rounds}: Spieler 1 hat folgendes gewürfelt: {self.player1.dice_roll}")
        print(f"Runde {self.rounds}: Spieler 2 hat folgendes gewürfelt: {self.player2.dice_roll}")
        print(f"Runde {self.rounds}: Spieler 3 hat folgendes gewürfelt: {self.player3.dice_roll}")

    def distribute_points(self):
        if self.player1.dice_roll == self.player2.dice_roll and self.player1.dice_roll == self.player3.dice_roll:
            print(f"Runde {self.rounds}: Alle haben die selbe Zahl gewürfelt, alle verlieren einen Punkt!")
            self.player1.lose_point()
            self.player2.lose_point()
            self.player3.lose_point()
        elif self.player1.dice_roll == self.player2.dice_roll or self.player1.dice_roll == self.player3.dice_roll or self.player2.dice_roll == self.player3.dice_roll:
            print(f"Runde {self.rounds}: Zwei Spieler haben die selbe Zahl gewürfelt, es werden keine Punkte vergeben!")
            return
        else:
            if self.player1.dice_roll > self.player2.dice_roll and self.player1.dice_roll > self.player3.dice_roll:
                self.player1.win_point()
                print(f"Runde {self.rounds}: Spieler 1 hat mit {self.player1.dice_roll} die höchste Zahl gewürfelt, er gewinnt einen Punkt!")
            elif self.player2.dice_roll > self.player3.dice_roll:
                self.player2.win_point()
                print(f"Runde {self.rounds}: Spieler 2 hat mit {self.player2.dice_roll} die höchste Zahl gewürfelt, er gewinnt einen Punkt!")
            else:
                self.player3.win_point()
                print(f"Runde {self.rounds}: Spieler 3 hat mit {self.player3.dice_roll} die höchste Zahl gewürfelt, er gewinnt einen Punkt!")

    def print_points(self):
        print(f"Runde {self.rounds}: Spieler 1 Punktestand: {self.player1.points}")
        print(f"Runde {self.rounds}: Spieler 2 Punktestand: {self.player2.points}")
        print(f"Runde {self.rounds}: Spieler 3 Punktestand: {self.player3.points}")

    def end_game(self):
        if self.player1.points == 6:
            print(f"Spieler 1 gewann das Spiel in der {self.rounds}. Runde!!!")
        elif self.player2.points == 6:
            print(f"Spieler 2 gewann das Spiel in der {self.rounds}. Runde!!!")
        else:
            print(f"Spieler 3 gewann das Spiel in der {self.rounds}. Runde!!!")
        self.reset_all()

    def reset_all(self):
        self.player1.reset()
        self.player2.reset()
        self.player3.reset()
        self.rounds = 0

my_game = Game()
my_game.start_game()

```

### a) Division durch die Listen Elemente

- die Zahl `1` soll durch das jeweilige Listenelement `numbers` dividiert werden. 
<br>**Achtung** hierbei soll der Code nicht unterbrochen werden und ein Fahler als Notiz ausgeworfen.
- numbers = [1.5, 3, 2, 0, 5]
- gebe jeweils das `Listenelement` und das `Rechenergebnis` an

```python
numbers = [1.5,3,2,0,5]

for num in numbers:
    try:
        print(f"1 / {num} = {1 / num}")
    except ZeroDivisionError as e:
        print(f"Error: {e}")
```

    - Gebe als Error den ´IndexError´ als catch indikator an mit dem Text: `Die Liste hat kein Element an dem Index`
