## Aufgabe 3: erstelle eine Funktion column_auswahl, welche einen Dataframe mit allen bestimmten Columnnamen erstellt
- Was sind die einzelnen Schritte welche ich hierfür brauche?
### a) Der neue Dataframe soll nur mit den Columns Newsletter enthalten sein
- in der Parameterliste soll zum einen der Dataframe und der Variablenname enthalten sein nach welchen Column gefiltert werden soll
- die KUNDENNUMMER ist der INDEX für den neuen Dataframes
- es soll keine Doubletten im neuen Dataframes enthalten sein
<br>
*299 rows × 48 columns*

```python
def column_auswahl(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    df = df.copy()
    # 1. Spalten auswählen, die den Suchbegriff enthalten
    columns = [col for col in df.columns if column_name in col]

    # 2. Subframe erstellen
    result = df[columns + ["KUNDENNUMMER"]]

    # 3. KUNDENNUMMER als Index setzen
    result = result.set_index("KUNDENNUMMER")

    # 4. Duplikate entfernen
    result = result[~result.index.duplicated()]

    return result

test = column_auswahl(data_basis, "Newsletter")
test
```

### b) erweitere bzw passe die Funktion column_auswahl
- um den Parameter in_out --> dieser gibt an ob ein Dataframes erstellt wir nur mit Newslettern oder alle anderen 
<br>
out = 25079 rows × 11 columns
<br>
in = 299 rows × 48 columns

```python
# True für bleibt drin, False für fällt raus
def column_auswahl2(df: pd.DataFrame, column_name: str, in_out: bool) -> pd.DataFrame:
    df = df.copy()
    columns: list

    if in_out:
        # Newsletter rein
        columns = [c for c in df.columns if column_name in c]
    else:
        # Newsletter raus
        columns = [c for c in df.columns if column_name not in c and c != "KUNDENNUMMER"]


    # 2. Subframe erstellen
    result = df[columns + ["KUNDENNUMMER"]]

    # 3. KUNDENNUMMER als Index setzen
    result = result.set_index("KUNDENNUMMER")

    # 4. Duplikate entfernen
    result = result[~result.index.duplicated()]

    # 5 Unnamed entfernen
    result = result.loc[:, ~result.columns.str.startswith("Unnamed")]

    return result

test = column_auswahl2(data_basis, "Newsletter", False)
test
```

## Aufgabe 4: insert column by condition

- erstelle die Funktion insert_columns_after_condition
- nach Newsletter_12_Cupon_20 soll eine Column mit dem dem gleichen Columnnamen wie davor mit dem Zusatz _count hinzugefügt werden
- ändere die Variablen in dem Dataframes von ja zu 1.0 und von nein zu 0.

```python
def insert_columns_after_condition(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    target_col = "Newsletter_12_Cupon_2020"

    df = df.replace({"ja": 1.0, "nein": 0.0})

    idx = df.columns.get_loc(target_col)

    # 3. neue Spalte erzeugen (z. B. identischer Inhalt)
    new_col_name = f"{target_col}_count"
    new_col_values = df[target_col]

    # 4. Spalte exakt nach target_col einfügen
    df.insert(idx + 1, new_col_name, new_col_values)

    return df

insert_columns_after_condition(column_auswahl2(data_basis, "Newsletter", True))
```

### c) prepare_every_second_column
schriebe eine Funktion prepare_every_second_column

```python
# ich denke mal er meint einfach jede zweite column umbenennen
# ich gehe davon aus, dass die erste spalte den index 0 hat also ist die zweite spalte ungerade
def prepare_every_second_column(df: pd.DataFrame) -> pd.DataFrame:
    result = df.copy()

    result.columns = [f"{col}_ungerade" if i % 2 != 0 else col for i, col in enumerate(result.columns)]

    return result

test = prepare_every_second_column(data_basis)
test
```
