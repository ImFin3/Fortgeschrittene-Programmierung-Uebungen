# Baum-Suche

```python
from collections import deque

baum_strukture_liste = {
    'A': ['B', 'C'],    # Wurzel
    'B': ['D', 'E'],    # 1 Ebene
    'C': ['F', 'K'],    # 1 Ebene --> K ist ein neuer Zweig unter C
    'D': ['G'],         # 2 Ebene
    'E': ['J'],         # 2 Ebene --> J ist ein neues Blatt unter E
    'F': ['H'],         # 2 Ebene
    'G': ['I'],         # 2 Ebene --> I ist die tiefste Ebene (Ebene 3) unter G
    'H': [],
    'I': [],
    'J': [],
    'K': []
}

baum_strukture_liste
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.children = []  # Liste statt left/right, da ein Knoten viele Kinder haben kann

# --- DEIN DICTIONARY ---
baum_struktur_liste = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'K'],
    'D': ['G'],
    'E': ['J'],
    'F': ['H'],
    'G': ['I'],
    'H': [], 'I': [], 'J': [], 'K': []
}

# --- SCHRITT 1: Alle Knoten-Objekte erstellen ---
# Wir erstellen ein neues Dictionary, das die Buchstaben (Strings) 
# mit den echten Node-Objekten verknüpft.
knoten_objekte = {name: Node(name) for name in baum_struktur_liste}

# --- SCHRITT 2: Die Kinder verknüpfen ---
# Wir gehen durch das ursprüngliche Dictionary und fügen jedem 
# Objekt seine Kinder-Objekte hinzu.
for eltern_name, kinder_namen in baum_struktur_liste.items():
    eltern_node = knoten_objekte[eltern_name]
    for kind_name in kinder_namen:
        kind_node = knoten_objekte[kind_name]
        eltern_node.children.append(kind_node)

# Jetzt ist 'A' ein echtes Objekt mit Kindern, die wiederum Kinder haben.
A = knoten_objekte['A']
```

```python
class Node:
    def __init__(self, value):
        self.value = value        # Speichert z.B. "A"
        self.left_child = None    # Platzhalter für das linke Kind-Objekt
        self.right_child = None   # Platzhalter für das rechte Kind-Objekt
        
# Schritt A: Erstelle die einzelnen Knoten-Objekte
A = Node('A')
B = Node('B')
C = Node('C')
D = Node('D')
E = Node('E')
F = Node('F')
G = Node('G')
H = Node('H')
I = Node('I')
J = Node('J')
K = Node('K')

# Schritt B: Verknüpfe sie (Identisch mit deiner Liste)
# Wurzel Ebene
A.left_child = B
A.right_child = C

# Ebene 1
B.left_child = D
B.right_child = E

C.left_child = F
C.right_child = K

# Ebene 2
D.left_child = G  # D hat nur ein linkes Kind laut Struktur
E.left_child = J  # E hat nur ein linkes Kind

F.left_child = H  # F hat nur ein linkes Kind

# Ebene 3 (Die tiefste Ebene)
G.left_child = I
```

## a) für eine Liste

Die BFS arbeitet nach dem Warteschlangen-Prinzip (Queue: First-In, First-Out).

1) Initialisierung: 
    - Erstellen Sie eine leere Liste für die besuchten Knoten ***und*** 
    - eine queue (Warteschlange), in der anfangs nur der Startknoten liegt. ``deque``
2) Die Schleife läuft Solange noch jemand in der Warteschlange steht.
3) Entnehmen: Nimm den vordersten (ältesten) Knoten aus der Schlange (``popleft``).
4) Prüfen: Wenn dieser Knoten noch nicht besucht haben:
    - Füge ihn der besucht-Liste hinzu.
    - Nimm alle seine Nachbarn (Kinder) und stelle sie hinten an die Warteschlange an (extend).
5) Ergebnis: Am Ende gib die Liste der besuchten Knoten zurück.

## b) Objekte

Objektstruktur, bei der jeder Knoten genau zwei Nachfolger hat: einen linken und einen rechten.

- Zuerst muss geprüft werden, ob der Startknoten (root) überhaupt existiert. Wenn nicht, wird eine leere Liste zurückgegeben.
- Die Warteschlange (Queue): Die queue speichert die Knoten, die noch besucht werden müssen. Sie folgt dem Prinzip First-In-First-Out (FIFO).
- Der Loop: Die while-Schleife läuft so lange, wie sich Knoten in der Warteschlange befinden.
    - Entnahme (pop(0)): In jedem Schritt wird der älteste Knoten (der am weitesten links/oben in der Schlange steht) herausgenommen.
    - Speichern: Der Wert dieses Knotens wird der Ergebnisliste result hinzugefügt. Hierfür brauch es for der while-Schleife eine leere Liste
- Expansion: Falls der aktuelle Knoten Kinder hat (left_child oder right_child), werden diese hinten an die Warteschlange angehängt. Dadurch werden sie erst verarbeitet, wenn alle Knoten der aktuellen Ebene fertig sind.

# 2) Tiefensuche

## a) Listen-Suche
Die DFS nutzt oft Rekursion. Das bedeutet, die Funktion ruft sich selbst für jedes Kind auf.
1) Vorbereitung: Da die Funktion sich selbst aufruft, müssen wir eine Liste ``besucht`` durch alle Aufrufe "mitreichen", 
    - welche ``None`` ist --> Beim allerersten Aufruf erstellen wir diese Liste neu
    - mit dem Namen ``dfs_order``
2) Fügen den aktuellen Knoten sofort der besucht-Liste hinzu.
3) Abstieg: Gehe jedes Kind des aktuellen Knotens durch.
4) Rekursion: Wenn ein Kind noch nicht besucht wurde, rufe die dfs_order-Funktion direkt für dieses Kind auf.
5) Rückgabe: Gebe am Ende die fertige Liste zurück

## b) Objektsuche

- eine Rekursion, welche prüft, ob wir an einem "leeren" Ende (einem Blatt ohne Kinder) angekommen sind., sonst läuft es Endlos
- Zuerst ruft sich die Funktion selbst für das linke Kind auf. Das Programm springt sofort so tief wie möglich in den linken Ast.
- Erst wenn der komplette linke Ast fertig abgearbeitet ist, kehrt das Programm zurück und führt den Befehl für das rechte Kind aus.
