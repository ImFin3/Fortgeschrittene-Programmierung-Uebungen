### 2c) Interpolationssuche

Anforderungen:
- lege die Startbedingungen fest
    - bestimme den Suchbereich und definiere eine untere Grenze ``low`` am Anfang der Liste (Index 0) ``und`` 
    - eine obere Grenze ``high`` am Ende der Liste

- Die Such-Schleife 
    -  suchen so lange weiter, wie die zwei Bedingungen erfüllt sind:
        - Der Suchbereich ist noch gültig (die untere Grenze ist nicht über die obere hinausgewandert).
        - Das gesuchte Element (target) liegt wertmäßig überhaupt innerhalb der aktuellen Grenzen (es ist größer oder gleich dem kleinsten Wert und kleiner oder gleich dem größten Wert des aktuellen Bereichs).

- Die intelligente Positionsschätzung
    - Anstatt stur die Mitte zu wählen, berechnen  eine geschätzte Position (mid). Dabei nutzen wir eine mathematische Formel, die schaut, wie weit der Zielwert vom kleinsten Wert entfernt ist, und dies ins Verhältnis zur gesamten Spannweite der Werte im aktuellen Suchbereich setzt.
    - **Idee:** Wenn die Zahlen in der Liste von 10 bis 1000 gehen und  die 20 gesucht wird, wird die Schätzung Sie sehr nah an den Anfang der Liste führen.
    - Formel: `mid = low + (target - arr[low]) * (high - low) // (arr[high] - arr[low])`
        - Erklärung: 
            - **(target - arr[low])** misst den "Abstand" des Zielwerts vom kleinsten Wert im aktuellen Suchbereich
            - **(high - low) // (arr[high] - arr[low])**  schauen, wie viele Index-Schritte pro Wert-Einheit zur Verfügung stehen. Wir teilen die Anzahl der verfügbaren Plätze (Indizes) durch die Spanne der Werte.

- Überprüfung der Schätzung
    - Treffer: Wenn der Wert an dieser Position genau unser Ziel ist, haben wir es gefunden und geben diese Position zurück.
    - Zu klein: Ist der Wert an der geschätzten Stelle kleiner als unser Ziel, wissen wir, dass das Ziel weiter rechts liegen muss. Wir verschieben unsere untere Grenze (low) auf die Position direkt rechts von unserer aktuellen Schätzung
    - Zu groß: Ist der Wert dort größer als unser Ziel, muss es links liegen. Wir verschieben unsere obere Grenze (high) auf die Position direkt links von unserer Schätzung.

- Wird die Schleife beendet, ohne dass wir das Element gefunden haben (weil der Suchbereich leer ist oder das Element außerhalb der Werte-Spannweite liegt), geben wir einen Standardwert wie -1 zurück. Dies signalisiert, dass das Element in der Liste nicht existiert.

## 5) Implementieren ein Buble Sort
- Anforderung:
    - schreibe eine Funktion namens ``bubbleSort``, die eine Liste als Parameter akzeptiert
    - Vergleiche das erste und das zweite Element, wenn das erste Element größer als das zweite ist, tausche beide aus.
    - Gehe zum nächsten Paar, d. h. zum zweiten und dritten, und mache dasselbe.
    - Wiederhole diesen Vorgang, bis das Ende des Arrays erreicht ist
    - Wiederhole alle oben genannten Schritte n-1 Mal
    <br><br>
- Ermittele zuerst die Gesamtlänge der Liste (n)
- die Liste muss mehrfach durchlaufen werden. Erstellen Sie eine Schleife, die insgesamt n-mal läuft. Nach jedem dieser großen Durchläufe ist sichergestellt, dass mindestens eine weitere Zahl (die jeweils größte des verbleibenden Bereichs) ihre endgültige Position am rechten Rand gefunden hat.
- Innerhalb jedes Durchlaufs benötigen Sie eine zweite Schleife (nasty). Diese wandert von links nach rechts durch die Liste.
- ***Wichtig***: Da nach jedem großen Durchlauf am Ende der Liste bereits die richtigen (großen) Zahlen stehen, muss diese innere Schleife jedes Mal ein Stück kürzer laufen (bis zum Index n−i−1)
- Der Tausch-Mechanismus (Swap)
    - Vergleiche innerhalb der inneren Schleife das aktuelle Element (lst[j]) mit seinem direkten Nachbarn rechts (lst[j+1]).
    - Die Bedingung: Wenn die linke Zahl größer ist als die rechte, vertauschen Sie die beiden Plätze. In Python können Sie dies effizient in einer Zeile lösen: a, b = b, a.
