### b) Basis - Durchschnittsberechnung 
- berechne den Durchschnitt aus der Liste `1,2,3,4,5,6,7,8,9` mit einer `Lambda Funktion` 
- erstelle eine Liste mittels for loop
- erstelle eine Lambda Funktion
     - Durchschnitt ist die Summe durch die Anzahl der Elemente
- gebe sowohl die Liste als auch das Durchschnittsrgebnis mittels print aus

## Aufgabe 2: map
### a) Listenelement * 4
- erstelle eine Liste von `1 bis 20` in der aber nur jedes `zweite Element` erstellt wird
- multipliziere jeden einzelnen Listenwert mit 4
- erstelle eine Liste mit einem for Loop

```python
list_times_four = []

for i in range(1,21,2):
    list_times_four.append(i)

result = [x * 4 for x in list_times_four]

for item in result:
    item *= 4

print(result)
```

- nun `löse` die Aufgabenstellung aus der Celle vorher mit einer `lambda Funktion`
- **Tipp:** map muss auch verwendet werden

```python
result = list(map(lambda x: x *4, list_times_four))

print(result)
```

```python
my_list = []

for i in range(1,10):
    my_list.append(i)

list_average = sum(my_list) / len(my_list)

list_delta = [value - list_average for value in my_list]

print(f"neue Daten = {list_delta}")
```

- nutze für die Aufgabe die kombination aus `map und lambda` Funktion

```python
my_list = []

for i in range(1,10):
    my_list.append(i)

list_average = average(my_list)

list_delta = list(map(lambda x: x - list_average, my_list))
print(f"map und filter = {list_delta}")
```

## Aufgabe 3: filter
### a) Basis
- erstelle eine Liste mit dem Namen `random_list`

- gebe alle Werte aus der Liste `random_list` welche zwischen `1 und 6` aus
    - speichere die Werte zwischen `1 und 6` in einer neuen Liste mit dem Namen `neue_liste`
    - **Tipp:** erstelle die Liste `neue_liste` mit einem for Loop, welche diese Werte angibt

```python
random_list = [random.randint(-4,10) for i in range(20)]
new_list = []
for i in random_list:
    if 1 <= i <= 6:
        new_list.append(i)
print(random_list)
print(new_list)
```

- verkürzte Schreibweise mit einer list-comprehension für die Aufgabe aus der Celle davor

```python
new_list = [x for x in random_list if 1 <= x <= 6]
print(new_list)
```

- nutze für die Aufgabe aus der Celle davor eine Kombination aus der `filter-lambda` Funktion

```python
new_list = list(filter(lambda x: 1 <= x <= 6, random_list))
print(new_list)
```

## Aufgabe 5: groupby
- Datenbasis ist der Dateframe `data_groupby`
- gebe das summierte Ergebnis für die Regionen aus `data_groupby` an für alle Columns

```python
# numeric only, weil die summe von strings keinen Sinn ergibt, man könnte es aber auch machen
temp = data_groupby.groupby("REGION").sum(numeric_only=True)
temp
```

- gebe das summierte Ergebnis für die Regionen an
- für die  Spalten wie `LISTENPREIS, VERKAUFSPREIS, RABATT_PREIS, EINKAUFPREIS, VK_PERIS_IN_USD`

```python
temp = (data_groupby[["LISTENPREIS","VERKAUFSPREIS","RABATT_PREIS","EINKAUFPREIS","VK_PERIS_IN_USD","REGION"]]
        .groupby("REGION")
        .sum(numeric_only=True))

temp
```

- zähle alle Bestellungen, welche ein Kunde getätigt hat 
- speicher das Ergebnis ist der Variabel data_groupby_count
<br> length: 299

```python
data_groupby_count = data_groupby.groupby("KUNDENNUMMER").size()
data_groupby_count
```

- summiere die Verkaufpreise pro Kunden und Getränkeart
- speicher diese unter der Variabel `data_groupby_verkauf`
<br> 1195 rows

```python
data_groupby_verkauf = data_groupby.groupby(["KUNDENNUMMER", "GETRAENK"])["VERKAUFSPREIS"].sum()
data_groupby_verkauf
```

- füge die beiden Variabeln `data_groupby_verkauf und data_groupby_count` in einen neuen Dataframe `data` zusammen 
- verändere anschliessend den Columnnamen `BESTELLNUMMER` in `Count`
<br><br>
1195 rows × 4 columns

```python
# erstmal die Series objekte in dataframes wandeln indem man den index resetted
data_groupby_count = data_groupby_count.reset_index(name="BESTELLNUMMER")
data_groupby_verkauf = data_groupby_verkauf.reset_index()

# dann mergen
data = data_groupby_verkauf.merge(
    data_groupby_count,
    on="KUNDENNUMMER",
    how="left"
)
data = data.rename(columns={"BESTELLNUMMER": "Count"})
data
```

- bestimme den median pro GETRAENK 
- füge diesen an den Dataframe `data` unter den Columnnamen `MEDIAN` an <br><br>
- Tipp: nutze transform
<br><br>1195 rows × 5 columns

```python
data["MEDIAN"] = (
    data
        .groupby("GETRAENK")["VERKAUFSPREIS"]
        .transform("median")
)
data
```

- gebe den minimalen `VERKAUFSPREIS pro KUNDENNUMMER` in einer neuen Column wieder
<br><br>
1195 rows × 5 columns

```python
data["MIN_VERKAUFSPREIS"] = data.groupby("KUNDENNUMMER")["VERKAUFSPREIS"].transform("min")
data
```

- sortiere den Dataframe `data_groupby`
    - wobei die `KUNDENNUMMER` als erstes sortiert werden soll und zwar `aufsteigend` 
    - die zweite Sortiercolumn soll der `VERKAUFSPREIS` sein, diese soll `absteigend` sortiert werden soll
- erstelle einen sortierten Sub-Dataframes `data_sort` mit den Columns `KUNDENNUMMER, VERKAUFSPREIS, GETRAENK und REGION`
- erstelle eine neue Column `Kundenbestellungen_count`
    - in der Column soll pro Kundennummer fortlaufend alle zuvor sortierten Bstellungen gezählt werden beginnend bei 1
<br><br> **Tipp: cumcount()**
<br> 25079 rows × 5 columns

```python
data_sort = data_groupby[["KUNDENNUMMER", "VERKAUFSPREIS", "GETRAENK", "REGION"]].sort_values(["KUNDENNUMMER", "VERKAUFSPREIS"], ascending=[True, False])

data_sort["KUNDENBESTELLUNGEN_COUNT"] = (
    data_sort
        .groupby("KUNDENNUMMER")
        .cumcount()
        + 1
)

data_sort
```

- erstelle einen neuen Dataframes mit einem beliebigen namen
    - Die Bedingung für den neuen Dataframes ist, dass nur pro Kunde die 60 grössten Verkaufpreise beinhaltet sind
    <br>
<br> 16459 rows × 5 columns

```python
data_top60 = data_sort[
    data_sort["KUNDENBESTELLUNGEN_COUNT"] <= 60
]

data_top60
```
